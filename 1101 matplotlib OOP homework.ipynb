```python
% matplotlib notebook
class Option :
    def __init__(self, K, Interval, Premium_Call, Premium_Put,):
        import numpy as np
        import matplotlib.pyplot as plt
        self.K = K
        self.Interval = Interval
        self.Premium_Call = Premium_Call
        self.Premium_Put = Premium_Put
        #self.ST = np.arange ( K - Interval, K + Interval )  # Vectorization
    
    def Call (self) :
        import numpy as np
        import matplotlib.pyplot as plt   
        ST = np.arange(self.K - self.Interval, self.K + self.Interval)
        Payoff_LongCall = np.maximum (ST - self.K, 0) - self.Premium_Call   # Long Call
        Payoff_ShortCall = - Payoff_LongCall
        plt.figure
        plt.plot (ST,Payoff_LongCall)
        plt.plot (ST,Payoff_ShortCall)
        plt.plot(ST, Payoff_LongCall, label=u'$+C$' ,color="red",linewidth=1) # 加上label參數添加圖例
        plt.plot(ST, Payoff_ShortCall, label=u'$-C$',color="blue",linewidth=1)
        plt.xlabel("$Index$", size = 20)
        plt.ylabel('$Payoff$', size = 20,rotation=0)
        plt.legend(loc = 'best')
        x0 = 9279
        y0 = 0
        plt.scatter(x0,y0,s=50,color='k')
        plt.annotate(r'$%s$' % x0,xy = (x0,y0),xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))
        plt.show()
        plt.axhline (0, color='black', ls='dashed')
        plt.axvline (K, color='black', ls='dashed')
        plt.axvline (K + Premium_Call, color='black', ls='dashed' )
    
    def Put(self):
        import numpy as np
        import matplotlib.pyplot as plt
        ST = np.arange(self.K - self.Interval, self.K + self.Interval) #S0~ST期 X軸
        Payoff_LongPut = np.maximum(K - ST, 0) - Premium_Put
        Payoff_ShortPut = - Payoff_LongPut
        plt.plot(ST, Payoff_LongPut)
        plt.plot(ST, Payoff_ShortPut)
        plt.plot(ST, Payoff_LongPut, label=u'$+P$' ,color="red",linewidth=1) # 加上label參數添加圖例
        plt.plot(ST, Payoff_ShortPut, label=u'$-P$',color="blue",linewidth=1)
        plt.xlabel("$Index$", size = 20)
        plt.ylabel('$Payoff$', size = 20,rotation=0)
        plt.legend(loc = 'best')
        x1 = 8914
        y1 = 0
        plt.scatter(x1,y1,s=50,color='k')
        plt.annotate(r'$%s$' % x1,xy = (x1,y1),xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))
        plt.axvline(K, color='black' ,ls='dashed')
        plt.axvline(K - Premium_Put, color='black', ls='dashed')
        plt.axhline(0, color='black', ls='dashed')   #0虛線
        
```
